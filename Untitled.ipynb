```python
def transformCSV(data):
    df = data[['carId','trackId', 'binIndex', 'lap_number', 'lap_distance', 'velocity_X', 'lap_time', 'fuel',  'world_position_X', 'world_position_Y', 'world_position_Z', 'throttle', 'brake']]
    df = df[df['lap_number'] > 0]
    df['velocity_X'] = df['velocity_X'] * 3.6
    df['binDistance'] = 1 # every bin is 1 meter on track, to get the total track this needs to be summed to a total
    df['totalDistance'] = df['binDistance'].cumsum() / 1000
    df['Time'] = pd.to_datetime(df.lap_time, unit='s').dt.time.astype(str).str[3:]
    data = df[['carId', 'trackId', 'binIndex', 'lap_number', 'lap_distance', 'velocity_X', 'lap_time', 'Time', 'fuel', 'totalDistance', 'world_position_X', 'world_position_Y', 'world_position_Z', 'throttle', 'brake']]
    return data
```
